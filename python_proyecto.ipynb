```python
# 1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias
# de cada letra en la cadena. Los espacios no deben ser considerados


def dic_frec(texto):
    texto_sin_espacios = texto.replace(" ","").lower()  # Formateamos el texto para homogeneizar y quitar espacios blancos

    frecuencias = {}

    for caracter in texto_sin_espacios: # Por cada caracter miramos una condicion if
        if caracter in frecuencias: # Si el caracter ya se encuentra en nuestro diccionario sumamos un numero mas en esa letra
            frecuencias[caracter] += 1  

        else: frecuencias[caracter] = 1  # Se añade un uno al diccionario para esa letra

    return frecuencias

prueba = dic_frec("holaa")
prueba    
```

```python
# 10.  Escribe una función que reciba una lista de números y calcule su promedio. Si la lista está vacía, lanza una 
# excepción personalizada y maneja el error adecuadamente.

# Creamos una excepcion nueva personalizada por si la lista esta vacia
class ListaVaciaError(Exception):
    pass # Pasa a la siguiente parte del codigo


def promedio(lista):

        if not lista:
            raise ListaVaciaError('La lista está vacia')
        
        else:
            # Promedio = suma(todos los valores) / (nº de valores)
            num_valores = len(lista) # nº de valores
            suma = sum(lista) # Suma los valores de la lista
            
            resultado = suma / num_valores

            return resultado

# Ahora vamos a manejar errores
try: 
    prueba = promedio([1,1,2,3,5])
    print(prueba)
except ListaVaciaError as e:
    print(e)
```

```python
# 34. Crea la clase Arbol , define un árbol genérico con un tronco y ramas como atributos. Los métodos disponibles son: 
# crecer_tronco , nueva_rama , crecer_ramas , quitar_rama e info_arbol . El objetivo es implementar estos métodos para manipular la estructura del árbol.

class Arbol:
    def __init__(self, tronco=1, ramas=None): # No podemos poner la lista vacia como parametro porque si no todos los arboles comparten la misma lista
        
        self.tronco = tronco
        self.ramas = [] if ramas is None else ramas # Hacemos esto para que cada arbol tenga su propia lista

        # Es lo mismo que:
        """
        if ramas = None:
            self.ramas = []
        else:
            self.ramas = ramas
        """

    def crecer_tronco(self):
        self.tronco += 1
    
    def nueva_rama(self):
        self.ramas.append(1)

    def crecer_ramas(self):
        self.ramas = list(map(lambda x: x+1, self.ramas))
    
    def quitar_rama(self, posicion):
        self.ramas.pop(posicion)

    def info_arbol(self):
        return f"Información sobre el árbol: \n Longitud del tronco: {self.tronco} \n Número de ramas: {len(self.ramas)} \n Longitud de ramas: {self.ramas}"

# Crear arbol
arbol1 = Arbol()
# Crecer arbol en una unidad
arbol1.crecer_tronco()
# Nueva rama 
arbol1.nueva_rama()
# Crecer rama
arbol1.crecer_ramas()
# 2 Nuevas ramas
arbol1.nueva_rama()
arbol1.nueva_rama()
# Quitar rama en posicion 2
arbol1.quitar_rama(2) # La posicion 2 es la tercera rama ya que empieza a contar en 0
# Obtener informacion
print(arbol1.info_arbol())

```

```python
# 36 Crea la clase UsuarioBanco ,representa a un usuario de un banco con su nombre, saldo y si tiene o no cuenta 
# corriente. Proporciona métodos para realizar operaciones como retirar dinero, transferir dinero desde otro usuario y 
# agregar dinero al saldo

class UsuarioBanco:
    def __init__(self, nombre, saldo, cuenta):
        self.nombre = nombre
        self.saldo = saldo
        self.cuenta = cuenta

    def retirar_dinero(self, retiro):
        if self.cuenta == True:
            if retiro > 0 and retiro < self.saldo:
                self.saldo = self.saldo - retiro
                return f"Has retirado {retiro}, su saldo actual es: {self.saldo}"
            else:
                return "Excede la cantidad que puedes retirar"
        else:
            return "Necesita crear una cuenta de banco para realizar operaciones"
        
    def transferir_dinero(self, destinatario, cantidad): ## Queda por hacer esto
        if self.cuenta == True and destinatario.cuenta == True:

            if cantidad > 0 and self.saldo >= cantidad:
                self.saldo = self.saldo - cantidad
                destinatario.saldo = destinatario.saldo + cantidad
                return f"Se han transferido {cantidad} de {self.nombre} a {destinatario.nombre}"
            
            else: 
                return "No se puede enviar esa cantidad"          

        else:
            return "Necesita crear una cuenta de banco para realizar operaciones"


    def agregar_dinero(self, agregar):
        if self.cuenta == True:
            if agregar > 0:
                self.saldo = self.saldo + agregar
                return f"Has agregado {agregar}, su saldo actual es: {self.saldo}"
            else:
                return f"No se puede agregar esa cantidad"
        else:
            return "Necesita crear una cuenta de banco para realizar operaciones"

# Crear usuarios
bob = UsuarioBanco('bob', 50, True)
alicia = UsuarioBanco('alicia', 100, True)

# Agregar 20 a BoB
bob.agregar_dinero(20)
bob.agregar_dinero(20) # Le mando otros 20 porque si no se le puede enviar dinero a alicia

# Hacer transferencia de 80 desde bob a alicia
bob.transferir_dinero(alicia,80)

# Retirar 50 unidades de saldo a "Alicia"
alicia.retirar_dinero(50)
```

```python
# 37. Crea una función llamada procesar_texto que procesa un texto según la opción especificada: 
# reemplazar_palabras , procesar_texto .contar_palabras , eliminar_palabra . Estas opciones son otras funciones que tenemos que definir primero y llamar dentro 
# de la función 

# Creamos las funciones independientes
def contar_palabras(texto):
    texto_f = texto.lower().split()
    frecuencia={}

    for valor in texto_f:
        if valor not in frecuencia:
            frecuencia[valor] = 1
        else:
            frecuencia[valor] +=1
            
    return frecuencia


def reemplazar_palabras(texto, palabra_original, palabra_nueva):
    return texto.replace(palabra_original, palabra_nueva)

def eliminar_palabra(texto, palabra):
    palabras = texto.split()
    palabras_filtradas = [p for p in palabras if p != palabra]
    return ' '.join(palabras_filtradas)

# Creamos la funcion que va a llamar a las otras
def procesar_texto(texto, opcion, *args):
    if opcion == "contar":
        return contar_palabras(texto)
    
    elif opcion == "reemplazar":
        # args[0] = palabra_original, args[1] = palabra_nueva
        return reemplazar_palabras(texto, args[0], args[1])
    
    elif opcion == "eliminar":
        # args[0] = palabra a eliminar
        return eliminar_palabra(texto, args[0])

    else:
        return "Opción no válida"


prueba = procesar_texto("Hola me llamo Alejandro Hola", 'contar')
print(prueba)

prueba2 = procesar_texto("Hola me llamo Alejandro Hola", 'reemplazar', 'Hola', 'ADIOS')
print(prueba2)

prueba3 = procesar_texto("Hola me llamo Alejandro Hola",'eliminar','Alejandro')
print(prueba3)

```

```python
# 41. En este ejercicio, se te pedirá que escribas un programa en Python que utilice condicionales para determinar el 
# monto final de una compra en una tienda en línea, después de aplicar un descuento. El programa debe hacer lo siguiente:

# 1. Solicita al usuario que ingrese el precio original de un artículo.
# 2. Pregunta al usuario si tiene un cupón de descuento (respuesta sí o no).
# 3. Si el usuario responde que sí, solicita que ingrese el valor del cupón de descuento.
# 4. Aplica el descuento al precio original del artículo, siempre y cuando el valor del cupón sea válido (es decir, mayor a cero). Por ejemplo, descuento de 15€. 
# 5. Muestra el precio final de la compra, teniendo en cuenta el descuento aplicado o sin él. 
# 6. Recuerda utilizar estructuras de control de flujo como if, elif y else para llevar a cabo estas acciones en tu programa de Python.

precio_original = int(input("Escriba el precio original del producto: "))
cupon = input("¿Tienes cupón de descuento? Respuesta: si o no: ").lower()

if cupon == "si":
    descuento = int(input("Introduce el valor del cupón de descuento: "))
    print(f"El descuento es de: {descuento}")

    if descuento <= 0:
        print("No se puede aplicar ese cupón")

    elif descuento > precio_original:
        print(f"El precio original es {precio_original}, después del descuento se queda en: 0")

    else:
        precio_final = precio_original - descuento
        print(f"El precio original es {precio_original}, después del descuento se queda en: {precio_final}")

else: 
    print(f"Como no tiene cupón de descuento, el precio final es {precio_original}")
```
